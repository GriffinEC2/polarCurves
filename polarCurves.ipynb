```python
import numpy as np
import matplotlib.pyplot as plt
pi = np.pi
```

```python
theta = np.linspace(-pi, pi, 100)
r = abs(np.cos(theta))
#print(r)
fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
ax.plot(theta, r)
ax.set_rticks([0, 0.5, 1, 1.25])
ax.set_rlabel_position(-22.5)
plt.show()
```

```python
theta = np.linspace(0, 2*pi, 100)
r = abs(np.cos(theta))
#print(r)
plt.polar(theta, r, color='blue', linewidth=2)
ax.set_rticks([0, 0.5, 1, 1.5, 2])
ax.set_rlabel_position(-22.5)
plt.show()
```

```python
theta = np.linspace(0, 2*pi, 100)
r = abs(np.cos(theta))
#print(r)
x = r * np.cos(theta)
y = r * np.sin(theta)
#print(x)
plt.plot(x,y)
plt.title("Polar Curve (Cartesian)", color='white', fontsize=14)


plt.xlabel("x", color = 'white')
plt.ylabel("y", color = 'white')
plt.tick_params(axis='both', colors='white')
```
